```python
def get_speeches_per_speaker(
    df: pd.DataFrame, include_shared: bool = True, exclude: list[str] | None = []
):
    """
    Get the number of speeches per speaker.
    If include_shared is True, it will include speeches shared by multiple speakers.

    Args:
        df: The dataframe to analyze. It must have a 'speaker' column.
        include_shared: Whether to include speeches shared by multiple speakers.
            If True, it'll parse the 'speaker' column as a list of speakers and add it
            to each speakers count.
        exclude: List of speakers to exclude from the analysis.

    Returns:
        A series with the number of speeches per speaker.
    """
    if include_shared:
        df = pd.DataFrame(
            df["speaker"].map(
                lambda x: list(map(str.strip, x.split(",")))
                if isinstance(x, str)
                else x
            )
        )
        df = df.explode("speaker")

    if exclude:
        df = df[~df["speaker"].isin(exclude)]

    return df["speaker"].value_counts()
```

# Visualizar discursos de candidatos a lo largo del tiempo


```python
df.head()
```

```python
# parse dates
df['date'] = df['date'].apply(lambda x: pd.to_datetime(x))
df.head()
```

```python
N = 10 
top_N_speakers = get_speeches_per_speaker(df, exclude=["???", "Democratic Candidates", "Multiple Speakers"]).head(N).index

df_speeches = df[[ 'speaker', 'type', 'date' ]].copy()
df_speeches = df_speeches[df_speeches['speaker'].isin(top_N_speakers)]
df_speeches.head()
```

```python
df_speeches['month'] = df_speeches['date'].dt.month
df_speeches['week'] = df_speeches['date'].dt.isocalendar().week
df_speeches.head()
```

```python
speech_by_month = df_speeches.groupby(['speaker', 'month']).size().unstack(fill_value=0)
speech_by_month
```

```python
speaches_by_month_df = speech_by_month.stack().reset_index()
speaches_by_month_df.columns = ['speaker', 'month', 'count']
speaches_by_month_df = speaches_by_month_df.sort_values(['count', 'speaker', 'month'], ascending=[False, True, True])
speaches_by_month_df
```

```python
month_list = [
    "Enero",
    "Febrero",
    "Marzo",
    "Abril",
    "Mayo",
    "Junio",
    "Julio",
    "Agosto",
    "Septiembre",
    "Octubre",
]
plt.figure(figsize=(15, 8))
sns.lineplot(data=speaches_by_month_df, x="month", y="count", hue="speaker", linewidth=4)

plt.grid()
plt.xticks(list(range(1, 11)), labels=month_list, fontsize=12)
plt.yticks(fontsize=12)
plt.xlabel("Mes", fontsize=16)
plt.ylabel("Cantidad de discursos", fontsize=16)
plt.title("Cantidad de discursos por mes y por persona", fontsize=18)
plt.legend(fontsize=14)
plt.tight_layout()


plt.show()
```

```python
# Group by speaker and week number
speech_by_week = df_speeches.groupby(['speaker', 'week']).size().unstack(fill_value=0)
speech_by_week
```

```python
speech_by_week_df = speech_by_week.stack().reset_index()
speech_by_week_df.columns = ['speaker', 'week', 'count']
speech_by_week_df = speech_by_week_df.sort_values(['count', 'speaker', 'week'], ascending=[False, True, True])
speech_by_week_df
```
